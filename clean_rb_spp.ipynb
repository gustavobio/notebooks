```python
import pandas as pd
import numpy as np
import geopandas
import matplotlib.pyplot as plt
import re

spp = pd.read_csv("~/data/speciesLink_all_112419_20190513152918.txt", sep = "\t", quotechar = "\"", names = list(range(0,51)), skipinitialspace = True, skiprows = 1)
spp_header = pd.read_csv("~/data/speciesLink_all_112419_20190513152918.txt", nrows = 1, sep = "\t", header = None)
spp.columns = spp_header.loc[0]
rb = pd.read_csv("/home/gustavo/data/occurrence.txt", sep = "\t", skiprows = 1, header = None, names = list(range(0,44)))
rb_header = pd.read_csv("/home/gustavo/data/occurrence.txt", nrows = 1, sep = "\t", header = None, names = list(range(0,44)))
rb.columns = rb_header.loc[0]
reflora = pd.read_csv("/home/gustavo/data/rb_full.csv")
rb.columns.sort_values()
reflora.columns.sort_values()
rb.drop(['license', 'associatedOccurrences', 'datasetName'], axis = 1, inplace = True)
reflora.drop(['identificationRemarks', 'verbatimEventDate', 'identificationVerificationStatus', 'eventRemarks', 'bibliographicCitation'], axis = 1, inplace = True)
rb = rb.append(reflora)
spp.drop(['boundingbox', 'class', 'continentocean', 'datelastmodified', 'kingdom', 'maximumdepth', 'maximumelevation', 'minimumdepth', 'minimumelevation', 'ordem', 'phylum', 'preparationtype', 'previouscatalognumber', 'relatedcatalogitem', 'relationshiptype', 'sex', 'timeofday'], axis = 1, inplace = True)
rb.drop(['maximumElevationInMeters', 'minimumElevationInMeters', 'modified'], axis = 1, inplace = True)
rb.columns = rb.columns.str.lower()
rb.rename(columns = {'decimallatitude': 'latitude', 'decimallongitude': 'longitude', 'recordedby': 'collector', 'day': 'daycollected', 'month': 'monthcollected', 'year': 'yearcollected', 'fieldnotes': 'notes', 'scientificnameauthorship': 'scientificnameauthor', 'recordnumber': 'collectornumber'}, inplace = True)


def ranker(e):
    if (e == "var.") | (e == "VARIEDADE"):
        return "var."
    elif (e == "subsp.") | (e == "SUB_ESPECIE"):
        return "subsp."
    else:
        return np.nan

def joiner(df):
    df.taxonrank = df.taxonrank.apply(ranker)
    return df.apply(lambda x: ' '.join(x.fillna('')), axis=1).replace("[\s+]", " ").str.strip()

rb.scientificname = joiner(rb[['genus', 'specificepithet', 'taxonrank', 'infraspecificepithet']])

pd.concat([spp, rb]).to_csv("/home/gustavo/data/all.csv", index = False)
```
